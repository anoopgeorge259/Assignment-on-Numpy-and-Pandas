###Q1. Employee Salary Adjustment
A company stores salaries in a NumPy array.
salaries = np.array([45000, 52000, 61000, 48000, 75000])
Tasks:
1. Increase all salaries by 10%.
2. Find the average salary.
3. Find the maximum and minimum salary.
4. Print salaries greater than 60000.

```python
salaries = np.array([45000, 52000, 61000, 48000, 75000])
print("Original Salaries:", salaries)

# 1. Increase all salaries by 10%
updated_salaries = salaries * 1.10
print("\n1. Salaries after 10% increase:", updated_salaries)

# 2. Find the average salary
average_salary = np.mean(updated_salaries)
print("\n2. Average Salary:", average_salary)

# 3. Find maximum and minimum salary
max_salary = np.max(updated_salaries)
min_salary = np.min(updated_salaries)
print("\n3. Maximum Salary:", max_salary)
print("   Minimum Salary:", min_salary)

# 4. Print salaries greater than 60000
high_salaries = updated_salaries[updated_salaries > 60000]
print("\n4. Salaries greater than 60000:", high_salaries)
```

##Q2. Sales Data Analysis
Daily sales of a store for 10 days:
sales = np.array([120,150,180,200,170,160,210,220,190,175])
Tasks:
1. Compute total sales.
2. Compute mean and standard deviation.
3. Identify days where sales exceeded the average.
4. Normalize the sales using Min-Max scaling.

```python
sales = np.array([120,150,180,200,170,160,210,220,190,175])
print("Original Sales Data:", sales)

# 1. Total sales
print("\n1. Total Sales:", np.sum(sales))

# 2. Mean and Standard Deviation
mean_sales = np.mean(sales)
std_sales = np.std(sales)
print("\n2. Mean Sales:", mean_sales)
print("   Standard Deviation:", std_sales)

# 3. Days where sales exceeded average
above_avg_sales = sales[sales > mean_sales]
above_avg_days = np.where(sales > mean_sales)[0] + 1
print("\n3. Sales greater than average:", above_avg_sales)
print("   Days where sales exceeded average:", above_avg_days)

# 4. Min-Max Normalization
normalized = (sales - np.min(sales)) / (np.max(sales) - np.min(sales))
print("\n4. Normalized Sales:", normalized)
```

##Q3. Matrix Operations for Business Data
A company has sales data for 3 products across 4 regions.
sales_matrix = np.array([
[120,140,130,150],
[80,90,100,95],
[200,210,190,205]
])
Tasks:
1. Find total sales per product.
2. Find total sales per region.
3. Identify the best performing product.
4. Identify the region with highest sales.

```python
sales_matrix = np.array([[120,140,130,150],[80,90,100,95],[200,210,190,205]])
print("Sales Matrix:\n", sales_matrix)

# 1. Total sales per product (row-wise)
product_sales = np.sum(sales_matrix, axis=1)
print("\n1. Total Sales per Product:", product_sales)

# 2. Total sales per region (column-wise)
region_sales = np.sum(sales_matrix, axis=0)
print("\n2. Total Sales per Region:", region_sales)

# 3. Best performing product
best_product = np.argmax(product_sales)
print("\n3. Best Performing Product: Product", best_product + 1)
print("   Sales of Best Product:", product_sales[best_product])

# 4. Best region
best_region = np.argmax(region_sales)
print("\n4. Best Performing Region: Region", best_region + 1)
print("   Sales of Best Region:", region_sales[best_region])
```

##Q4. Employee Dataset Exploration
Create a DataFrame:
data = {

'Experience':[2,5,7,6,3,4]
}
Tasks:
1. Convert this dictionary into a DataFrame.
2. Display basic statistics of the dataset.
3. Find the average salary per department.
4. Identify employees with experience greater than 4 years.
5. Add a new column Bonus = 10% of salary

```python
data = {
'Employee':['A','B','C','D','E','F'],
'Department':['HR','IT','IT','Finance','HR','Finance'],
'Salary':[50000,65000,70000,72000,52000,68000],
'Experience':[2,5,7,6,3,4]
}

# 1. Create DataFrame
df = pd.DataFrame(data)
print("1. Employee DataFrame:\n", df)

# 2. Basic statistics
print("\n2. Basic Statistics:\n", df.describe())

# 3. Average salary per department
print("\n3. Average Salary per Department:\n", df.groupby('Department')['Salary'].mean())

# 4. Employees with experience > 4
print("\n4. Employees with Experience > 4:\n", df[df['Experience'] > 4])

# 5. Add Bonus column
df['Bonus'] = df['Salary'] * 0.10
print("\n5. DataFrame with Bonus Column:\n", df)
```

##Q5. Missing Data Handling
Dataset:
data = {
'Name':['John','Anna','Peter','Linda','David'],
'Age':[28, np.nan, 35, 32, np.nan],
'Salary':[50000,54000,np.nan,58000,60000]
}
Tasks:
1. Detect missing values.
2. Replace missing Age with mean age.
3. Replace missing Salary with median salary.
4. Remove rows with any remaining missing values

```python
data = {
    'Name': ['John','Anna','Peter','Linda','David'],
    'Age': [28, np.nan, 35, 32, np.nan],
    'Salary': [50000,54000,np.nan,58000,60000]
}

# Create DataFrame
df = pd.DataFrame(data)
print("Original DataFrame:\n")
print(df)

# 1. Detect missing values
print("\n1. Missing Values in Dataset:\n")
print(df.isnull())

print("\nNumber of Missing Values per Column:\n")
print(df.isnull().sum())

# 2. Replace missing Age with mean age (NO inplace)
mean_age = df['Age'].mean()
df['Age'] = df['Age'].fillna(mean_age)

print("\n2. After Filling Missing Age with Mean:\n")
print(df)

# 3. Replace missing Salary with median salary (NO inplace)
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("\n3. After Filling Missing Salary with Median:\n")
print(df)

# 4. Remove any remaining missing values
df_cleaned = df.dropna()

print("\n4. Final DataFrame after removing remaining missing values:\n")
print(df_cleaned)
```

##Q6. Sales Performance Analysis
Dataset:
sales_data = {
'Region':['North','South','East','West','North','South','East','West'],
'Salesperson':['A','B','C','D','E','F','G','H'],
'Sales':[12000,15000,13000,17000,16000,14000,15500,16500]
}
Tasks:
1. Find total sales per region.
2. Identify the top salesperson
3. Compute average sales across regions.
4. Sort the dataset by sales descending.

```python
sales_data = {
    'Region': ['North','South','East','West','North','South','East','West'],
    'Salesperson': ['A','B','C','D','E','F','G','H'],
    'Sales': [12000,15000,13000,17000,16000,14000,15500,16500]
}

df = pd.DataFrame(sales_data)
print("Original DataFrame:\n")
print(df)

# 1. Total sales per region
total_sales_region = df.groupby('Region')['Sales'].sum()
print("\n1. Total Sales per Region:\n")
print(total_sales_region)

# 2. Identify the top salesperson
top_index = df['Sales'].idxmax()
top_salesperson = df.loc[top_index]

print("\n2. Top Salesperson:\n")
print(top_salesperson)

# 3. Average sales across regions
avg_sales_region = df.groupby('Region')['Sales'].mean()
print("\n3. Average Sales per Region:\n")
print(avg_sales_region)

# 4. Sort dataset by sales (descending)
sorted_df = df.sort_values(by='Sales', ascending=False)
print("\n4. Dataset Sorted by Sales (Descending):\n")
print(sorted_df)
```

##Q7. Customer Purchase Analysis
Dataset:
data = {
'CustomerID':[101,102,103,104,105,106,107],
'Age':[25,45,35,50,23,40,60],
'PurchaseAmount':[250,600,400,800,150,700,900],
'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

Tasks:
1. Find the average purchase amount by city.
2. Identify customers with purchase above overall average.
3. Find the city generating the highest revenue.
4. Create a column Age_Group:

    o <30 → Young

    o 30–50 → Adult

    o 50 → Senior
5. Find average purchase per age group.

```python
data = {
    'CustomerID':[101,102,103,104,105,106,107],
    'Age':[25,45,35,50,23,40,60],
    'PurchaseAmount':[250,600,400,800,150,700,900],
    'City':['Delhi','Mumbai','Delhi','Chennai','Mumbai','Delhi','Chennai']
}

df = pd.DataFrame(data)
print("Original DataFrame:\n")
print(df)

# 1. Average purchase amount by city
avg_city = df.groupby('City')['PurchaseAmount'].mean()
print("\n1. Average Purchase Amount by City:\n")
print(avg_city)

# 2. Customers with purchase above overall average
overall_avg = df['PurchaseAmount'].mean()
above_avg_customers = df[df['PurchaseAmount'] > overall_avg]
print("\n2. Customers with Purchase above Overall Average:\n")
print(above_avg_customers)

# 3. City generating highest revenue
city_revenue = df.groupby('City')['PurchaseAmount'].sum()
top_city = city_revenue.idxmax()

print("\n3. City with Highest Revenue:", top_city)
print("Revenue:", city_revenue[top_city])

# 4. Create Age_Group column
df['Age_Group'] = np.where(df['Age'] < 30, 'Young',
                   np.where(df['Age'] <= 50, 'Adult', 'Senior'))
print("\n4. DataFrame with Age_Group:\n")
print(df)

# 5. Average purchase per age group
avg_age_group = df.groupby('Age_Group')['PurchaseAmount'].mean()
print("\n5. Average Purchase per Age Group:\n")
print(avg_age_group)
```

Q8. Time Series Sales Analysis


Generate data:

dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date':dates,'Sales':sales})

Tasks:
1. Convert Date to datetime index.
2. Compute monthly percentage growth.
3. Calculate 3-month rolling average sales.
4. Identify the month with highest growth rate

```python
dates = pd.date_range('2024-01-01', periods=12, freq='M')
sales = [200,220,250,270,300,320,350,360,380,400,420,450]
df = pd.DataFrame({'Date': dates, 'Sales': sales})
print("Original DataFrame:\n")
print(df)

# 1. Convert Date to datetime index
df.set_index('Date', inplace=True)
print("\n1. DataFrame with Date as Index:\n")
print(df)

# 2. Monthly percentage growth
df['Growth (%)'] = df['Sales'].pct_change() * 100
print("\n2. Monthly Percentage Growth:\n")
print(df)

# 3. 3-month rolling average
df['Rolling Avg (3 months)'] = df['Sales'].rolling(window=3).mean()
print("\n3. 3-Month Rolling Average:\n")
print(df)

# 4. Month with highest growth rate
max_growth_date = df['Growth (%)'].idxmax()
max_growth_value = df['Growth (%)'].max()
print("\n4. Month with Highest Growth Rate:")
print("Date:", max_growth_date)
print("Growth (%):", max_growth_value)
```
